```python
import pandas as pd
import numpy as np
path = "./gdrive/MyDrive/cs464_hw2_csv_file/dataset.csv"
data = pd.read_csv(path)
data
```

```python
dataset = data.to_numpy()
```

```python
def normalize(data): # To normalize the dataset
    for i in range(np.shape(data)[1]): 
        indiced_data=data[:,i]
        x_max=np.max(indiced_data)
        x_min=np.min(indiced_data)
        normalized_data=(indiced_data-x_min)/(x_max-x_min)
        data[:,i]= normalized_data   
    return data

def sigmoid(x): # sigmoid function with having + in the denominator
  return 1/(1+np.exp(x))

def get_weight(mu=0,sigma=1,n=13):  # gets weight from gaussian dist
  return np.random.normal(mu,sigma, size=(n,1))

def MCLE(y,z):  # z is sum_of_wx_and_w0
  if y == 1:
    return 1-sigmoid(z)
  else:
    return sigmoid(z)

def update_weights(feature_vector, labels, learning_rate_alpha, prev_weights):
        sum_of_wx=np.dot(feature_vector, prev_weights[1:]) # gets sum of x_i * w_i
        z=prev_weights[0]+sum_of_wx # sum with w0
        probability_vector= MCLE(1,z) # gets a vector of probabilities from MCLE
        loss=labels-probability_vector # y-y_hat
        new_weights=np.zeros(prev_weights.shape[0])
        new_weights[1:] = prev_weights[1:] + learning_rate_alpha*(np.dot(np.transpose(feature_vector),loss))
        new_weights[0]=prev_weights[0]+learning_rate_alpha*np.sum(loss)        
        return new_weights
```
